05. 데이터 가져오기

```python
hospital = data["TbHospitalInfo"]["row"]
hospital
```

06. 필요한 정보 추출

```python
information_ls = [] # 병원 정보를 담을 빈 리스트
for i in hospital:
    information_dict = {} # 각 병원에 대해 새로운 딕셔너리를 하나 만들 것
    information_dict["기관명"] = i["DUTYNAME"]
    information_dict["병원분류명"] = i["DUTYDIVNAM"]
    information_dict["간이약도"] = i["DUTYMAPIMG"]
    information_dict["주소"] = i["DUTYADDR"]
    information_ls.append(information_dict) # 새로 생성된 딕셔너리를 리스트에 추가
information_ls
```

연습문제 | 추출한 정보에 전화번호 추가

```python
information_ls = [] # 병원 정보를 담을 빈 리스트
for i in hospital:
    information_dict = {} # 각 병원에 대해 새로운 딕셔너리를 하나 만들 것
    information_dict["기관명"] = i["DUTYNAME"]
    information_dict["병원분류명"] = i["DUTYDIVNAM"]
    information_dict["간이약도"] = i["DUTYMAPIMG"]
    information_dict["주소"] = i["DUTYADDR"]
    information_dict["대표전화1"] = i["DUTYTEL1"]
    information_ls.append(information_dict) # 새로 생성된 딕셔너리를 리스트에 추가
information_ls
```

03) pandas를 이용하여 정보 정리하기

01) pd.DataFrame: 딕셔너리를 데이터프레임으로 변환하는 클래스
02) information_ls: 딕셔너리들이 들어있는 리스트
-> information_ls 리스트 안의 딕셔너리들을 한 번에 표로 만들어 줄 것

```python
table = pd.DataFrame(information_ls)
table
```

04. 데이터를 표로 저장

```python
table.to_csv("./병의원위치정보.csv", encoding = "utf-8")
```
