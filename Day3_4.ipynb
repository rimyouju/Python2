```python
# tag_title의 부모 태그 출력
print(tag_title.parent)
```

[출력 순서]

#### 연습문제 

```python
from bs4 import BeautifulSoup
html = """
<html>
<head>
<title class = "main title" id = "t-id">Welcome to my page</title>
</head>
<body>
<div>
<p class = "desc">오늘은<span>파이썬</span>을 배우는 날입니다.</p>
<p class = "desc>HTML과<span>BeautifulSoup</span>을 함께 배워요</p>
</div>
</body>
</html>
"""
```

연습문제 01 | title 태그 전체와 텍스트 출력

```python
soup = BeautifulSoup(html, "lxml")
print(soup.title)
print(soup.title.text)
```

연습문제 02 | title 태그의 속성 딕셔너리 출력

```python
print(soup.title.attrs)
print(soup.title.get('class'))
print(soup.title["id"])
```

연습문제 03 | .text vs .string 차이 확인

```python
p__tag = soup.find("p", class_="desc")
print("text:", p__tag.text)
print("string:", p__tag.string)
```

연습문제 04 | 자식 요소 접근(contens, children)

```python
print("contents:", p__tag.contents)
for child in p__tag.children:
    print("child", child)
```

연습문제 05 | 부모 및 조상 요소 접근

```python
span_tag = soup.find("span")
print("부모 태그:", span_tag.parent.name)
```

```python
print("조상 태그:")
for parent in span_tag.parents:
    print("_", parent.name)
```
