2. 자동 로그인

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
# 1. 로그인 페이지 열기
driver.get("https://the-internet.herokuapp.com/login")
# 2. 아이디 입력
username = driver.find_element(By.ID, "username")
username.send_keys("tomsmith")

# 3. 비밀번호 입력
password = driver.find_element(By.ID, "password")
password.send_keys("SupperSecretPassword!")
# 4. 로그인 버튼 클릭
login_button = driver.find_element(By.CLASS_NAME, "radius")
login_button.click()
# 5. 결과 확인을 위해 잠시 대시
time.sleep(2)

# 6. 스크린샷 저장(로그인 후 화면)
driver.save_screenshot("login_result.png")
print("로그인 테스트 완료")
driver.quit() # 종료
```

3. webdriver가 하는 일

- driver.get(...) -> 브라우저에서 페이지 열기
- driver.find_element(...) -> 요소 찾기
- send_keys(...) -> 키보드 입력
- click() -> 클릭
- driver.save_screenshot(...) -> 스크린샷
- driver.quit() -> 브라우저 종료

4. 웹 페이지에서 특정 html 요소를 찾는 방법

driver.find_element(...) -> 요소 찾기

01) By.ID

ex) <input type = "text" id = "username"/>

-> driver.find_element(By.ID, "username")

2) By.CLASS_NAME

ex) <button class = "radius">Login</button>

driver.find_element(By.CLASS_NAME, "radius")

3) By.NAME

ex) <input type = "password" name = "password" />

driver.find_element(By.NAME, "password")

## 02 | 무한 스크롤 5번 내리기

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
# 1. 페이지 열기
driver.get("https://the-internet.herokuapp.com/infinite_scroll")
# 2. 페이지 로딩 기다리기
time.sleep(2)
# 3. 스크롤 5번 내리기
for i in range(5):
    # 자바스크립트를 이용해 현재 페이지의 맨 아래로 스크롤 이동
    # window : 현재 브라우저 창(탭)을 의미
    # .scrollTo(x, y): 브라우저 화면을 좌표 (x, y)로 이동
    # driver.body.scrollHeight: y축 좌표 = 페이지의 전체 높이,
    # 이 값을 넣으면 페이지의 가장 아래로 이동
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    print(f"스크롤 {i + 1}회 완료")
    time.sleep(2)
print("무한 스크롤 테스트 완료!")
driver.quit()
```
