02) urlopen()을 사용해서 웹사이트 요청
- 웹 페이지에 접속해서 HTML 문서를 가져오는 함수
- urlopen()을 사용해서 웹사이트에 요청을 보냄
- 받은 HTML 데이터를 BeautifulSoup으로 분석
- 원하는 태그나 정보를 꺼내서 사용 가능

```python
from urllib.request import urlopen
```

03) 가져오고 싶은 웹사이트 주소 작성

```python
url = "https://www.naver.com/"
```

04) 웹사이트에 접속해서 HTML 데이터를 가져옴
- urlopen() 함수는 웹사이트에 요청을 보내고, HTML 데이터를 받아오는 역할

```python
page = urlopen(url)
```

05) html.parser로 HTML 구조를 자동으로 읽고 분석

```python
soup = BeautifulSoup(page, "html.parser")
```

예제1: <title>태그 전체를 출력

```python
print(soup.title)
```

예제2: <title> 안의 텍스트만 출력

```python
print(soup.title.text)
```
