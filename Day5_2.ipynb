## 03 | 크롤링과 파일 입출력 예제

01) 파싱 후 텍스트 파일에 결과 저장

```python
from bs4 import BeautifulSoup
html = """
<!DOCTYPE html>
<html lang = "ko">
<head>
    <meta charset = "UTF-9"
    <title>크롤링 실습 페이지</title>
</head>
<body>
    <h1>크롤링 실습용 페이지입니다</h1>
    <p class = "description"> 이 페이지는 크롤링 연습을 위한 예제 입니다.</p>
    <ul id = "items">
        <li class = "item">사과</li>
        <li class = "item">바나나</li>
        <li class = "item">체리</li>
        <li class = "item">포도</li>
    </ul>
    <img src = "https://via.placeholder.com/150" alt = "예시 이미지">
    <a href = "https://example.com" target = "_blank">예제 링크</a>
</body>
</html>
"""
```

```python
soup = BeautifulSoup(html, 'html.parser')
title_tag = soup.find("h1")
items = soup.find_all("li", class_="item")
img_tag = soup.find("img")

# 화면 출력
print("제목:", title_tag.text)
print("리스트 항목:")
for item in items:
    print("-", item.text)
if img_tag:
    print("이미지 주소:", img_tag['src'])
else:
    print("이미지 태그가 없습니다.")

# 파일 생성
with open('crawling_file.txt', 'w', encoding = 'utf-8') as f:
    f.write(f"제목: {title_tag.text}\n")
    f.write("리스트 항목: \n")
    for item in items:
        f.write(f"- {item.text}\n")
    if img_tag:
        f.write(f"이미지 주소: {img_tag['src']}\n")
    else:
        f.write("이미지 태그가 없습니다.\n")
```

연습문제 | 헤드라인 뉴스 파일에 저장하기

```python
import requests
from bs4 import BeautifulSoup
url = "https://news.naver.com/section/105"
headers = {"User-Agent" : "Mozilla/5.0"}
response = requests.get(url, headers = headers)
# 응답 성공
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    title_tags = soup.select("div.sa_text strong.sa_text_strong")
    print("전체 뉴스 헤드라인 목록")
    for i, tag in enumerate(title_tags, 1):
        print(f"{i}. {tag.get_text(strip = True)}")
# 응답 실패 
else:
    print("요청 실패:", response.status_code)

# 파일 생성
with open('crawling_news_file.txt', 'w', encoding = 'utf-8') as f:
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        title_tags = soup.select("div.sa_text strong.sa_text_strong")
        f.write("전체 뉴스 헤드라인 목록")
        for i, tag in enumerate(title_tags, 1):
            f.write(f"{i}. {tag.get_text(strip = True)}\n")
    else:
        f.write("요청 실패:", response.status_code)
```
