### 산술연산에 사용하는 내장함수

- abs(x): x의 절댓값을 반환
- bool(x): x의 논리값(True 또는 False를 반환)
- divmod(a, b): a를 b로 나누었을 때의 몫과 나머지로 구성된 튜플을 반환
- float(x): 문자열 또는 수로 입력받은 x를 부동소수점 수로 변환하여 값을 반환, x를 생략하면 0.0을 반환
- hex(x): 정수값 xdml 16진수 문자열을 반환
- int(x): x를 int형 정수로 변환한 값을 반환
- max(arg1, arg2,,,): 인수의 최댓값을 반환
- min(arg1, arg2,,,): 인수의 최솟값을 반환
- oct(x): 정수값 x에 해당하는 8진수 문자열을 반환
- pow(x, y): x의 y 제곱인 (x**y)를 반환
- pow(x, y, z): x의 y제곱을 z로 나누었을 때의 나머지를 반환
-> pow(x, y) % z 보다 효율적으로 계산 가능
- round(n, digits): n의 소수부를 ndigits 자릿수가 되도록 반올림한 값을 반환
- sum(x, start): x의 원소값을 처음부터 끝까지 순서대로 더한 총합에 start값을 더하여 반환

```python
# abs(x) : 절댓값 반환
print(abs(-10))
```

```python
# bool(x) : 논리값 반환
print(bool(0))
print(bool(42))
print(bool(""))
print(bool("hi"))
print(bool([]))
```

```python
# divmod(a, b) : 몫과 나머지 반환
print(divmod(10, 3)) # (몫, 나머지)
```

```python
# float(x) : 실수 변환
print(float(5))
print(float("3.14"))
print(float())
```

```python
# hex(x) : 16진수 반환
print(hex(5))
```

```python
# int(x) : 정수 변환
print(int(3.14))
print(int("42"))
print(int("101", 2)) # 2진수를 10진수로 반환
```

```python
# max(ards) : 최댓값 반환
print(max(1, 2, 5, 7, 10))
print(max([3, 7, 8, 2]))
```

```python
# min(args) : 최솟값 반환
print(min(4, -1, 9, 0))
print(min([10, 2, 5, 9]))
```

```python
# oct(x) : 8진수 반환
print(oct(8))
print(oct(64))
```

```python
# pow(x, y) : 거듭제곱
print(pow(2, 3))
```

```python
# pow(x, y, z) : 거듭제곱 후 나머지 연산
print(pow(2, 3, 5)) # (2 ** 3) % 5
```

```python
# round(n, digits) : 반올림
print(round(3.14159, 2))
```

```python
print(round(2.5)) # 짝수 쪽으로 반올림
print(round(3.5))
```

```python
# sum(iterrable, start) : 합계 계산
numbers = [1, 2, 3, 4, 5]
print(sum(numbers))
```

```python
print(sum(numbers, 10)) # 합계 + 10
```
