3. 요청 헤더(봇 차단 방지용)
헤더란?
- User-Agent와 같은 HTTP 요청을 보낼 때 함께 보내는 정보
- "나 이런 환경에서 이걸 요청해! 브라우저는 이거고, 언어는 이거, 쿠키는 이런 걸 사용해!"

- F12 - Network = headers -> User-Agent: Mazilla/5.0

```python
headers = {"User-Agent" : "Mozilla/5.0"}
```

(ex) Accept-Language: 선호하는 언어를 서버에 알림
    Referer: 어디서 왔는지(이전 페이지 주소) -> Referer를 아예 설정하지 않으면 차단됨

headers = {
    "User-Agent" : "Mozilla/5.0",
    "Accept-Language" : "ko-KR, ko; q = 0.9",
    "Referer" : "주소"
}

4. GET 요청: "이 주소의 정보를 줘" 라고 요청하는 방법

```python
response = requests.get(url, headers = headers)
```

5. 응답이 성공적일 때
- 기사제목 가져오기
- soup.select().         
  : CSS 선택자로 HTML 요소를 리스트 형태로 반환.        
    조건에 맞는 요소가 여러 개면 전부 가져옴

```python
from bs4 import BeautifulSoup
html = """
<div class = "sa_text">
<strong class = "sa_text_strong"> 헤드라인1 </strong>
</div>
<div class = "sa_text">
<strong class = "sa_text_strong"> 헤드라인2 </strong>
</div>
"""

soup = BeautifulSoup(html, "html.parser")
result = soup.select("div.sa_text strong.sa_text_strong")
print(result)
```

```python
print(result[0])
```

```python
for content in result:
    print(content.text)
```

- get_text()는 태그 안의 텍스트만 가져옴
- strip = True는 양쪽 공백 제거

```python
for content in result:
    print(content.get_text(strip = True))
```

```python
# 5. 응답이 성공적일 때 -> 기사 제목 가져오기
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")

    title_tags = soup.select("div.sa_text strong.sa_text_strong")
    print("전체 뉴스 헤드라인 목록")
    for i, tag in enumerate(title_tags, 1): # enumerate 출력되는 것에 인덱스 번호를 첨부해서 출력
        print(f"{i}. {tag.get_text(strip = True)}")
# 6. 응답 실패시 
else:
    print("요청 실패:", response.status_code)
```

응답코드 알아보기

- 403: 접근 금지(봇 차단 등)
- 404: 페이지를 찾을 수 없음
- 500: 서버 내부 오류

연습문제 | IT/과학 헤드라인 가져오기

```python
import requests
from bs4 import BeautifulSoup
url = "https://news.naver.com/section/105"
headers = {"User-Agent" : "Mozilla/5.0"}
response = requests.get(url, headers = headers)
# 응답 성공
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    title_tags = soup.select("div.sa_text strong.sa_text_strong")
    print("IT 칼럼 헤드라인 목록")
    for i, tag in enumerate(title_tags, 1):
        print(f"{i}. {tag.get_text(strip = True)}")
# 응답 실패 
else:
    print("요청 실패:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
url = "https://news.naver.com/section/105"
headers = {"User-Agent" : "Mozilla/5.0"}
response = requests.get(url, headers = headers)
# 응답 성공
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    title_tags = soup.select("a.ss_text_headline")
    print("IT 칼럼")
    for i, tag in enumerate(title_tags, 1):
        print(f"{i}. {tag.get_text(strip = True)}")
# 응답 실패 
else:
    print("요청 실패:", response.status_code)
```
