02) time | 시간 관련 작업을 할 때 사용

01_ time 모듈의 time() 함수 | 현재 시간

```python
import time
now = time.time()
print("현재시간(초단위):", now) 
```

** time.time() 함수는
현재 시간을 1970년 1월 1일 0시 0초 이후 경과한 초로 반환

* 사람이 읽을 수 있는 시간으로 변경하려면
-> datetime 모듈을 사용
* datetime 모듈의 fromtimestamp() 함수
-> 초 단위 숫자를 사람이 읽을 수 있는 datetime 객체로 변환

datetime 모듈의 strtime() 함수
- string format time의 약자
- datetime의 객체를 문자열로 변환

```python
import time
from datetime import datetime

# 현재 시간
now = time.time()
print("현재 시간:", now)

# datetime 객체로 변환
dt = datetime.fromtimestamp(now)
print("datetime 객체:", dt)

# 포맷팅된 문자열로 출력
formatted = dt.strftime("%Y-%m-%d")
print("형식화된 시간:", formatted)
```

날짜 시간 관련 포맷 코드
- %Y 연도(4자리)
- %y 연도(2자리)
- %m 월(2자리)
- %d 일(2자리)
- %H 시(24시간제, 2자리)
- %M 분(2자리)
- %S 초(2자리)

```python
f = dt.strftime("%m")
print("월:", f)
```

연습문제 | 파일 입출력을 통한 프로젝트

파일에 내용 쓰고 읽는 프로그램 만둘기
- 사용자가 입력한 내용을 텍스트 파일에 저장하고, 저장된 파일 내용을 불러와 화면에 출력하는 간단한 프로그램
- 파일에 한 줄 쓰고, 읽어서 풀력하는 기능만 있는 예제

```python
# 쓰기 기능
def write_file():
    text = input("파일에 쓸 내용을 입력하세요:")
    with open("data.txt", "w", encoding = "utf-8") as f: # 파일을 f라는 객체명으로
        f.write(text)
        print("파일에 저장했어요!")

# 읽기 기능
def read_file():
    with open("data.txt", "r", encoding = "utf-8") as f: # 파일을 f라는 객체명으로
        content = f.read()
        print(content)

# 실행
while True:
    print('1. 쓰기, 2. 읽기, 3. 종료')
    select = int(input("선택:"))
    if select == 1:
        print(select)
        write_file()
    elif select == 2:
        print(select)
        try:
            read_file()
        except FileNotFoundError:
            print("파일이 없어요!")
    elif select == 3:
        print(select)
        print("끝!")
        break
    else:
        print(select)
        print("1, 2, 3 중에서 선택하세요.")
```

연습문제 | 일기 쓰기

1. 일기 쓰기
- 사용자가 입력한 내용을 텍스트 파일로 저장
- 파일명은 일기 작성 날짜로 자동 생성
- 내용은 UTF-8 인코딩으로 저장
- 저장 완료 후, 성공 메시지 풀력

2. 일기 읽기
- 사용자가 날짜(YYYY-MM-DD)를 입력하면, 해당 날짜 파일을 열어 내용을 출력
- 파일이 없으면 "해당 날짜의 일기가 존재하지 않습니다." 메시지 출력
- 읽기 실패 시 에러 메시지 출력

3. 프로그램 메뉴
- 프로그램이 종료될 때까지 아래 메뉴를 반복해서 보여줌
#### - 1. 일기 쓰기
#### - 2. 일기 읽기
#### - 3. 종료
- 잘못된 입력에는 다시 입력을 요청

- 날짜 및 시간 처리에 datetime 모듈 사용
- 파일 입출력은 open() 함수와 with 문을 활용하여 처리
- 예외 초리를 통해 오류 발생 시 적절한 안내 메시지를 출력
- 파일명은 YYYY-MM-DD.txt 형식을 엄수
- 프로그램 종료 시 메시지 출력

```python
import time
from datetime import datetime

# 일기 쓰기
def write_diary():
    text = input("일기를 작성하세요:")
    now = datetime.now().strftime("%Y-%m-%d")
    with open(now, "w", encoding = "utf-8") as f:
        f.write(text)
        print("저장 완료!")

# 일기 읽기
def read_diary():
    date = input("날짜를 입력하세요:")
    try:
        with open(date, "r", encoding = "utf-8") as f:
            content = f.read()
            print(content)
    except FileNotFoundError:
        print("해당 날짜의 일기가 존재하지 않습니다.")

while True:
    print("1. 일기 쓰기")
    print("2. 일기 읽기")
    print("3. 종료")
    select = int(input("선택:"))
    if select == 1:
        print(select)
        write_diary()
    elif select == 2:
        print(select)
        read_diary()
    elif select == 3:
        print(select)
        print("프로그램을 종료합니다.")
        break
    else:
        print(select)
        print("1, 2, 3 중에서 선택하세요.")
```
